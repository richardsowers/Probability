```python
import numpy
import datetime
%matplotlib inline
#%matplotlib notebook
import matplotlib.pyplot as plt
import pandas
import dateutil
import scipy.stats
```

```python
N=1000 #number of simulations
```

## Student t and Normal ##

```python
numpy.random.seed(0)
samples=scipy.stats.t.rvs(3,size=N)
print(samples[:10])
```

```python
plt.figure()
scipy.stats.probplot(samples, plot=plt)
plt.xlabel("normal quantiles")
plt.ylabel("student quantiles")
plt.show()
```

```python
ticker="GOOG"
data=pandas.read_csv(ticker+".csv",index_col=0)
print(data.head())
```

```python
prices=data["Adj Close"]
print(prices.head())
```

```python
log_returns=numpy.diff(numpy.log(prices))
print(log_returns[:10])
```

```python
plt.figure()
scipy.stats.probplot(log_returns, plot=plt)
plt.xlabel("normal quantiles")
plt.ylabel("return quantiles")
plt.title("QQ plot of returns for "+ticker)
plt.show()
```
